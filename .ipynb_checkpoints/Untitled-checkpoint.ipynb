```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
link = pd.read_csv("myConnections.csv")
link['Connected On'] = pd.to_datetime(link['Connected On'])
link.set_index('Connected On', inplace=True, drop=True)
link.sort_index(inplace=True)
link = link.assign(connected=1)
```

```python
link.head(1)
```

```python
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
FIGSIZE = (10, 7)
FONT = {"family": "Google", "weight": "normal", "size": 16}
tds = "#0073b1"
week = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

```python
def weekly_connection(link):
    weekly = link[["connected"]].resample("W").sum()

    fig, ax = plt.subplots(figsize=FIGSIZE)

    plt.plot(weekly.index, weekly.connected, c=tds)

    plt.title("Connection growth", fontdict=FONT, y=1.2)
    plt.ylabel("Weekly connections", fontdict=FONT)

    ax.set_frame_on(False)
    plt.grid(True)
    plt.show()
weekly_connection(link)
```

```python
def bar_column(df, col):
    fnames = df[col].value_counts().head(30)
    plot_fnames(fnames,col)
    
    

def plot_fnames(fnames, col, index="index"):
    fnames = fnames.reset_index()

    fig, ax = plt.subplots(figsize=FIGSIZE)

    plt.bar(
        x=fnames.index,
        height=fnames[col],
        color=tds,
        alpha=0.5
    )

    plt.title("{} distribution".format(col), fontdict=FONT, y=1.2)
    plt.xticks(
        fnames.index,
        fnames[index].str.capitalize(),
        rotation=65,
        ha="right",
        size=FONT["size"],
    )

    plt.ylabel("Occurence", fontdict=FONT)
    plt.yticks()#[0, 5, 10, 15, 20])
    ax.set_frame_on(False)
    plt.grid(True)

    plt.show()
```

```python
bar_column(link, "Company")
```

```python
bar_column(link, "Position")
```

```python
def nlp_cv(link):
    tfidf = CountVectorizer(ngram_range=(1, 3), stop_words='english')
    cleaned_positions = list(link["Position"].fillna(""))
    res = tfidf.fit_transform(cleaned_positions)
    res = res.toarray().sum(axis=0)

    fnames = pd.DataFrame(
        list(sorted(zip(res, tfidf.get_feature_names())))[-30:],
        columns=["Position by Words Freq", "Words"]
    )[::-1] 
    plot_fnames(fnames, "Position by Words Freq", "Words")

nlp_cv(link)
```
